```python
import numpy as np
import pandas as pd
import os
df = pd.read_csv("dream_market_cocaine_listings_edited.csv", engine="python")
```

```python
df.head()
```

```python
# carbon footprint from transportation
distance = pd.read_csv("distance.csv", engine="python")
distance.head()
```

```python
df = df.merge(distance, how="outer", on='ships_from_to')
df.head()
```

```python
ts = df.copy()
ts['carbon_footprint_transportation'] = 20 * ts['distance'] * 0.0000103 * ts['successful_transactions']
ts.head()
```

```python
route = ts.groupby(['ships_from_to'], as_index = False)['carbon_footprint_transportation'].sum()
route
```

```python
# carbon footprint from transportation (bia)
df = pd.read_csv("dream_market_cocaine_listings_edited.csv", engine="python")
distance_bia = pd.read_csv("distance_bia.csv", engine="python")
df = df.merge(distance_bia, how="outer", on='ships_from_to')

# distance_bia
ts_bia = df.copy()
ts_bia['carbon_footprint_transportation'] = 20 * ts_bia['distance'] * 0.0000103 * ts_bia['successful_transactions']
ts_bia
```

```python
route_bia = ts_bia.groupby(['ships_from_to'], as_index = False)['carbon_footprint_transportation'].sum()
route_bia
```
