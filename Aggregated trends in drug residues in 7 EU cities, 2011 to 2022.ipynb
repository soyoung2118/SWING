```python
import numpy as np
import pandas as pd
import os
df = pd.read_csv("ww-data-long-2023-05-02.csv", engine="python")
# cocaine, amphetamine, MDMA, methamphetamine, cannabis
```

```python
df.head()
```

```python
df = df.drop(['Wednesday',	'Thursday',	'Friday',	'Saturday',	'Sunday',	'Monday',	'Tuesday',	'Weekday mean',	'Weekend mean'], axis=1)
df.head()
```

```python
# 2011 to 2022, cocaine data in filtered_cities
df_cocaine = df[df['Metabolite'] == 'cocaine']
filtered_cities = ['Antwerp Zuid', 'Zagreb', 'Milan', 'Eindhoven', 'Utrecht', 'Castellon', 'Santiago']
df_cocaine = df_cocaine[df_cocaine['City'].isin(filtered_cities)]
df_cocaine
```

```python
# Aggregated trends in cocaine residues in 7 EU cities, 2011 to 2022
df_cocaine_mean = df_cocaine.groupby(['Year'], as_index = False)['Daily mean'].mean()
df_cocaine_mean
```

```python
from matplotlib import pyplot as plt
x_values = df_cocaine_mean['Year']
y_values = df_cocaine_mean['Daily mean']
plt.plot(x_values, y_values, label="Value (mg/1000people/day)")
plt.title('Aggregated trends in cocaine residues in 7 EU cities, 2011 to 2022')
plt.legend()
plt.show()
```
