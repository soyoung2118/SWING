```python
import pandas as pd
import matplotlib.pyplot as plt

# ships_from_to 데이터
data = {
    'ships_from_to': [
        'CA->WW', 'CN->WW', 'CZ->WW',
        'DE->WW', 'ES->EU',
        'ES->EU', 'ES->WW', 'EU->WW', 'FR->WW',
        'GB->EU', 'GB->WW', 'IT->WW', 'NL->WW', 'SE->EU', 'US->WW', 'US->WW', 'WW->WW'
    ],
    'total_gram': [
        761.25, 2000, 42129.348, 11960, 1723.333,
        70, 400, 14078.103, 2394.361, 7908,
        6726.029, 5223.429, 197811.429,
        1, 723.179, 236.25, 920
    ],
    'carbon_footprint': [
        0.05, 0.13, 0.05, 2.77, 0.05, 0.01, 0.03,
        0.93, 0.16, 0.5, 0.44, 0.34,
        13.02, 0, 0.05, 0.03, 0.06
    ]
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 선 그래프 생성
fig, ax1 = plt.subplots(figsize=(12, 6))
ax1.plot(df['ships_from_to'], df['total_gram'], marker='o', linestyle='-', label='Total Gram')
ax1.set_xlabel('Origin and Destination', fontsize=7)
ax1.set_ylabel('Total Gram(g)')

# 오른쪽 y축 설정
ax2 = ax1.twinx()
ax2.plot(df['ships_from_to'], df['carbon_footprint'], marker='o', linestyle='-', color='red', label='Carbon Footprint')
ax2.set_ylabel('Carbon Footprint(kg)')
ax2.set_ylim(0, 14)

# 그래프 제목 설정
plt.title('Total Gram and Carbon Footprint by Origin and Destination')

# 범례 표시
lines = [ax1.get_lines()[0], ax2.get_lines()[0]]
ax1.legend(lines, [line.get_label() for line in lines])

# 축 레이블 회전
plt.xticks(rotation=90)

# 그래프 출력
plt.tight_layout()
plt.show()
```
