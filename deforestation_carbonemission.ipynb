```python
import pandas as pd
import matplotlib.pyplot as plt

# 넓이 데이터
data = {
    'patch attribute': ['Honduras', 'Guatemala',  'Panamá', 'Nicaragua'],
    'Hectares': [102.0, 7979.4, 226.3, 179.7]
}

df = pd.DataFrame(data)

# 그래프 생성
plt.scatter(df.index, df['Hectares'], s=df['Hectares'], alpha=0.5)

# 축 레이블 및 제목 설정
plt.ylabel('Hectares')
plt.title('Area Representation')

plt.xticks([])

# 국가 이름 표시
for i, name in enumerate(df['patch attribute']):
    plt.text(i, df['Hectares'][i], name, ha='center', va='bottom')

plt.show()
```

```python
import matplotlib.pyplot as plt


data = {
    'patch attribute': ['Honduras',  'Guatamala', 'Panamá', 'Nicaragua'],
    'Min_year': [2004, 2001, 2001, 2002, 2001],
    'Max_year': [2012, 2013, 2012, 2012, 2013]
}

# 그래프 생성
for i in range(len(data['patch attribute'])):
    country = data['patch attribute'][i]
    forest_period = (data['Min_year'][i], data['Max_year'][i])
    drug_period = (2006, 2010)  # 마약 생산이 증가한 기간

    # 겹치는 부분 확인
    if forest_period[1] >= drug_period[0] and forest_period[0] <= drug_period[1]:
        intersect_start = max(forest_period[0], drug_period[0])
        intersect_end = min(forest_period[1], drug_period[1])

        # 겹치는 부분 표시
        plt.plot([intersect_start, intersect_end], [i+1, i+1], 'gray', linewidth=5)

    # 숲 황폐화 기간 표시
    plt.plot(forest_period, [i+1, i+1], 'b', linewidth=2)

    # 마약 생산 기간 표시
    plt.plot(drug_period, [i+1, i+1], 'r', linewidth=2)

# 축 설정
plt.ylim(0.5, len(data['patch attribute']) + 0.5)
plt.yticks(range(1, len(data['patch attribute'])+1), data['patch attribute'])
plt.xlabel('Year')
plt.title('Deforestation and Drug Production')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel('Global potential production of oven-dry opium in selected countries.xlsx')
df2 = pd.read_excel('라틴 아메리카 양귀비 경작지 데이터.xlsx')
data1 = df.iloc[2, 4:13]
data2 = df2.iloc[2, 4:13]


fig, ax1 = plt.subplots()

# 첫 번째 변수를 왼쪽 y축에 그림
ax1.plot(data1, color='blue', label='production')
ax1.set_xlabel('year')
ax1.set_ylabel('Ton', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# 두 번째 변수를 오른쪽 y축에 그림
ax2 = ax1.twinx()
ax2.plot(data2, color='red', label='cultivation')
ax2.set_ylabel('Hectar', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# 범례 표시
lines = [ax1.get_lines()[0], ax2.get_lines()[0]]
ax1.legend(lines, [line.get_label() for line in lines])

plt.title('connection between production and cultivation of oven-dry opium in selected countries')
plt.show()
```

```python
import matplotlib.pyplot as plt

df = pd.read_excel('Environmental impact results per crop.xlsx')


categories = ['cocaine', 'cacao', 'coffee', 'sugarcane']
rd_values = [1, 0.33, 0.74, 0.61]
po_values = [1, 0.05, 0.03, 0.02]
fae_values = [0.95, 0.92, 1, 1]
od_values = [1, 0.29, 0.76, 0.61]


x = range(len(categories))

plt.bar(x, rd_values, width=0.2, label='RD')
plt.bar(x, po_values, width=0.2, label='PO', bottom=rd_values)
plt.bar(x, fae_values, width=0.2, label='FAE', bottom=[rd + po for rd, po in zip(rd_values, po_values)])
plt.bar(x, od_values, width=0.2, label='OD', bottom=[rd + po + fae for rd, po, fae in zip(rd_values, po_values, fae_values)])


plt.xlabel('Products')
plt.ylabel('Values')
plt.title('Environmental Impact')


plt.legend()


plt.xticks(x, categories)

plt.show()
```
